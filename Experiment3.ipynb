```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate a dummy grayscale image for demonstration
# You can replace this with loading your own image, e.g., image = cv2.imread('/content/IMG_20251205_131833461.jpg', cv2.IMREAD_GRAYSCALE)
# image = np.zeros((200, 256), dtype=np.uint8)
# for i in range(256):
#     image[:, i] = i

# Or, uncomment the following line to use a real image from OpenCV samples (if available)
try:
    image = cv2.imread('/content/IMG_20251205_131833461.jpg', cv2.IMREAD_GRAYSCALE)
    if image is None:
        raise FileNotFoundError("Image not found. Generating a dummy image instead.")
except (cv2.error, FileNotFoundError) as e:
    print(f"Warning: {e}. Generating a dummy grayscale image for demonstration.")
    image = np.zeros((200, 256), dtype=np.uint8)
    for i in range(256):
        image[:, i] = i


# Display the original grayscale image and its histogram
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')

plt.subplot(1, 2, 2)
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
plt.plot(hist, color='black')
plt.title('Histogram of Original Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

```python
# 2. Perform Histogram Equalization
equalized_image = cv2.equalizeHist(image)

# Display the equalized image and its histogram
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.subplot(1, 2, 2)
hist_equalized = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])
plt.plot(hist_equalized, color='black')
plt.title('Histogram of Equalized Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```
