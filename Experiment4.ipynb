```python
import sys

def canny_edge_detection(image_path):
    """
    Performs Canny edge detection on an input image.

    Args:
        image_path (str): The path to the input image file.
    """
    # 1. Load the image from the specified path
    image = cv2.imread(image_path)

    # Check if image loading was successful
    if image is None:
        print(f"Error: Could not load image from {image_path}")
        return

    # 2. Convert the image to grayscale (Canny operates on grayscale images)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # 3. Apply Gaussian blur to reduce noise (optional but recommended)
    # Kernel size (5, 5), sigmaX=0 (auto-calculated)
    blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

    # 4. Perform Canny Edge Detection
    # Parameters: image, low_threshold, high_threshold
    # A common ratio for low:high threshold is 1:2 or 1:3
    low_threshold = 50
    high_threshold = 150
    edges = cv2.Canny(blurred_image, low_threshold, high_threshold)

    # 5. Display the original and edge-detected images
    # In a Colab environment, cv2.imshow requires a display server, which is usually not available.
    # To display images in Colab, you typically convert them to a format suitable for matplotlib or PIL.
    print("Canny edge detection performed.")
    print("Original image shape:", image.shape)
    print("Edges image shape:", edges.shape)
    # Example of how to display images in Colab using matplotlib:
    # from matplotlib import pyplot as plt
    # plt.figure(figsize=(10, 5))
    # plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
    # plt.subplot(122), plt.imshow(edges, cmap='gray'), plt.title('Canny Edges')
    # plt.show()
    return edges # Return the edges image for further processing

if __name__ == "__main__":
    # Use an image file name here
    input_image_name = '/content/465673.jpg' # Using an existing image file from the environment

    # Always use the hardcoded image name for Colab execution to avoid sys.argv issues
    canny_edge_detection(input_image_name)
```

```python
import cv2
import matplotlib.pyplot as plt

input_image_name = '/content/465673.jpg'

# Load the original image
original_image = cv2.imread(input_image_name)

if original_image is not None:
    # OpenCV loads images in BGR format, matplotlib expects RGB
    original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(10, 8))
    plt.imshow(original_image_rgb)
    plt.title('Original Image')
    plt.axis('off') # Hide axes ticks and labels
    plt.show()
else:
    print(f"Error: Could not load image from {input_image_name}")
```

```python
import matplotlib.pyplot as plt

# Assuming 'edges' is the variable holding the output from canny_edge_detection
# If canny_edge_detection was not run or its output was not assigned, this cell might fail.
# Let's re-run the canny_edge_detection and store its output to ensure 'edges' is available.
input_image_name = '/content/465673.jpg'
edges = canny_edge_detection(input_image_name)

if edges is not None:
    plt.figure(figsize=(10, 8))
    plt.imshow(edges, cmap='gray')
    plt.title('Canny Edge Detected Image')
    plt.axis('off') # Hide axes ticks and labels
    plt.show()
else:
    print("No edges image to display. Canny detection might have failed or returned None.")
```
